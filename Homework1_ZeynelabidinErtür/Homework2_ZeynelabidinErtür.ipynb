```python
# Question 3
first_pmf = thinkstats2.Pmf( school.agemomb1, label='school')
other_pmf = thinkstats2.Pmf(university.agemomb1, label='university')
thinkplot.PrePlot(2)
thinkplot.Hist(first_pmf, align='right', width=5)
thinkplot.Hist(other_pmf, align='left', width=5)
thinkplot.Config(xlabel='Value', ylabel='Percentage')
```
