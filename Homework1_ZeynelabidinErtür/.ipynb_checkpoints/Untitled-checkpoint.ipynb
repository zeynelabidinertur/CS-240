```python
x = resp.fmarno.value_counts()
x
```

```python
pist = {}
counter = 0
for i in x:
    pist[counter] = i
    counter = counter + 1
```

```python
pist
```

```python
hist = thinkstats2.Hist(pist, label='Number of Mariage Frequency')
thinkplot.Hist(hist, width = 0.5)
thinkplot.Config(ylabel='Number of Respondents', xlabel='Number of Mariage')
```

```python
neverMarried = resp[resp["fmarno"]==0]
married = resp[resp["fmarno"]!=0]
```

```python
neverMarried
```

```python
married
```

```python
marriedHist = {}

for m in married.totincr.value_counts().keys():
    marriedHist[m] = married.totincr.value_counts()[m]
```

```python
neverMarriedHist = {}

for m in neverMarried.totincr.value_counts().keys():
    neverMarriedHist[m] = neverMarried.totincr.value_counts()[m]
```

```python
marriedHist
```

```python
neverMarriedHist
```

```python
thinkplot.PrePlot(2)
hist1 = thinkstats2.Hist(marriedHist, label='Married')
hist2 = thinkstats2.Hist(neverMarriedHist, label='Never Married')
thinkplot.Hist(hist2, align='left', width = 0.5)

thinkplot.Hist(hist1, align='right', width = 0.5)
thinkplot.Config(ylabel='Number of Respondents', xlabel='Income')
```

```python
print("FOR MARRIED RESPONDENTS: \nMin Income: "+ str(married.totincr.min()) +"\nMax Income: "+ str(married.totincr.max())+"\nStandard Deviation of Income: "+ str(married.totincr.std())+"\nVariance of Income: "+ str(married.totincr.var())+"\nMean of Income: "+ str(married.totincr.mean()))
```

```python
print("FOR NEVER MARRIED RESPONDENTS: \nMin Income: "+ str(neverMarried.totincr.min()) +"\nMax Income: "+ str(neverMarried.totincr.max())+"\nStandard Deviation of Income: "+ str(neverMarried.totincr.std())+"\nVariance of Income: "+ str(neverMarried.totincr.var())+"\nMean of Income: "+ str(neverMarried.totincr.mean()))
```
