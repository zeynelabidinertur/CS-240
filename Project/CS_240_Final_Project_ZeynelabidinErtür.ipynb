# Section 2 | start

```python
import pandas as pd
import numpy as np
import thinkstats2
import thinkplot
master = pd.read_csv("basketball_master.csv") # importing the dataset which include height variable
all_star = pd.read_csv("basketball_player_allstar.csv") # importing the dataset which include player stats
```

```python
# cleaning the dataset to avoid invalid values like height==0 or height==NaN
master = master[(master[['height', 'weight']] != 0).all(axis=1)]
master = master[np.isfinite(master['height'])]
master = master[np.isfinite(master['weight'])]
```

```python
# cleaning the dataset to avoid unnecessary columns
all_star = all_star.drop(['conference', 'league_id'], axis=1)
# cleaning the dataset to avoid invalid values and dropping the columns which includes NaN as value
all_star = all_star[(all_star[['points']] <= 50).all(axis=1)]
all_star = all_star.dropna(axis='columns')
```

```python
# merging releated datasets by using player_id's 
merged_data = pd.merge(master, all_star, left_on="bioID", right_on="player_id")
```

```python
# taking the releated columns from cleaned and merged dataset and take mean of this values 
# to avoid duplicated rows and round that value to make graphs more understandable  
merged_data = merged_data.groupby('player_id')['height', 'points', 'rebounds', 'fg_made', 'assists', 'ft_made'].mean().round()
merged_data
```

```python
# dividing the merged dataset to categories players according to their heights
# hp = higher_players, mp = middle_players, sh = shorter_players | according to all players height mean
hp = merged_data.loc[merged_data['height'] > merged_data.height.mean()+merged_data.height.std()]
mp = merged_data.loc[merged_data['height'] < merged_data.height.mean()+merged_data.height.std()]
mp = mp.loc[mp['height'] > merged_data.height.mean()-merged_data.height.std()]
sp = merged_data.loc[merged_data['height'] < merged_data.height.mean()-merged_data.height.std()]
```

```python
# test class created
class DiffMeansPermute(thinkstats2.HypothesisTest):

    def TestStatistic(self, data):
        group1, group2 = data
        test_stat = abs(group1.mean() - group2.mean())
        return test_stat

    def MakeModel(self):
        group1, group2 = self.data
        self.n, self.m = len(group1), len(group2)
        self.pool = np.hstack((group1, group2))

    def RunModel(self):
        np.random.shuffle(self.pool)
        data = self.pool[:self.n], self.pool[self.n:]
        return data
```

```python
# in this section I have applied the test for every combination for each variable
h_s_points_data = hp.points, sp.points
hypotest1 = DiffMeansPermute(h_s_points_data)
pvalue = hypotest1.PValue()
pvalue
```

```python
# and plot the results
hypotest1.PlotCdf()
thinkplot.Config(xlabel='test statistic',
                   ylabel='CDF',title="Higher vs Shorter Points")
```
